```python
import pandas as pd
```

```python
df = pd.read_csv('../data/netflix_titles.csv')
```

```python
df.head(5)
df.info()
df.isnull().sum()
```

```python
#Преобразование в datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')

#Создание колонки под жанры & континента страны  
df_genres = df.assign(genre=df['listed_in'].str.split(',')) \
                    .explode('genre').assign(genre=lambda d: d['genre'].str.strip())

df_countries = df.assign(country=df['country'].str.split(','))\
                    .explode('country').assign(country=lambda d: d['country'].str.strip())
```

```python
df_genres.to_csv('../data/processed/genre.csv', index=False)
df_countries.to_csv('../data/processed/countries.csv', index=False)
df.to_csv('../data/processed/main.csv', index=False)
```
