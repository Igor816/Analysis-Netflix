```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('..\data\processed\main.csv')
df2 = pd.read_csv('..\data\processed\genre.csv')
df2.head(2)
```

```python
count_genre = df2.genre.value_counts().sort_values(ascending=False)
order = count_genre.index
```

```python
#   

plt.figure(figsize=(12, 8))          #   
ax = sns.countplot(
    y='genre',                         #  
    data=df2,
    order=order,
    palette='tab20c'                  #   
)

ax.set_xlabel(' ')
ax.set_ylabel('')
ax.set_title('     ')
plt.tight_layout()
plt.show()
```

```python
#  
year_count = df['release_year'].value_counts().reset_index()
px.line(year_count, x='count', y='release_year', title="    ").show()
```
