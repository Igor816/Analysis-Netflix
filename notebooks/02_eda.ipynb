```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import streamlit as st
```

```python
df = pd.read_csv('..\data\processed\main.csv')
df2 = pd.read_csv('..\data\processed\genre.csv')
df2.head(2)
```

```python
count_genre = df2.genre.value_counts().sort_values(ascending=False)
order = count_genre.index
```

```python
#График распределения по жанрам

plt.figure(figsize=(12, 8))          # высота‑ширина подбираются вручную
ax = sns.countplot(
    y='genre',                         # горизонтальная ось
    data=df2,
    order=order,
    palette='tab20c'                  # любой набор цветов
)

ax.set_xlabel('Количество фильмов')
ax.set_ylabel('Жанр')
ax.set_title('Количество фильмов по каждому из жанров')
plt.tight_layout()
plt.show()
```
